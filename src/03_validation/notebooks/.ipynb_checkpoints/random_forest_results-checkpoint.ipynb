```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV
df = pd.read_csv('../results/random_forest_grid_search_results.csv')  # Replace 'your_file.csv' with the actual file name

# Step 2: Find the Best Model for Each Feature Type
best_models = df.loc[df.groupby('features')['accuracy'].idxmax()]

# Step 3: Plot the Best Scores
# Setting the figure size
plt.figure(figsize=(10, 6))

# Creating a bar plot for accuracy, micro_f1, and macro_f1 for each feature type
width = 0.2  # Width of the bars
x = range(len(best_models))  # X positions for the groups

plt.bar([a - width for a in x], best_models['accuracy'], width=width, label='Accuracy')
plt.bar(x, best_models['micro_f1'], width=width, label='Micro F1')
plt.bar([a + width for a in x], best_models['macro_f1'], width=width, label='Macro F1')

# Adding labels and title
plt.xlabel('Feature Type')
plt.ylabel('Score')
plt.title('Best Model Scores by Feature Type')
plt.xticks(x, best_models['features'], rotation=90)  # Set the x-ticks to be the feature names

# Adding a legend
plt.legend()

# Showing the plot
plt.tight_layout()  # Adjust layout to not cut off labels
plt.show()
```
