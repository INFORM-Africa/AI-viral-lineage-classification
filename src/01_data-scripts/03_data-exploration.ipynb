```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_parquet('../../data/processed/genomes.parquet', engine='pyarrow')  # You can use 'fastparquet' as the engine
```

```python
# Count the frequency of different categories in "Lineage"
lineage_counts = data['Lineage'].value_counts()

# Plot
plt.figure(figsize=(30, 6))
lineage_counts.plot(kind='bar')
plt.title('Frequency of Different Categories in "Lineage"')
plt.xlabel('Lineage')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot a histogram of the percentages
plt.figure(figsize=(10, 6))
plt.hist(data['Percentage'], bins=20, alpha=0.7)
plt.title('Histogram of Percentage of Specified Characters in Sequences')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()
```
