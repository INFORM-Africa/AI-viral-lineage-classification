```python
import numpy as np
import os
import glob
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, matthews_corrcoef
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
def find_ancestors(label):
    ancestors = []
    parts = label.split('.')
    for i in range(len(parts)):
        ancestors.append('.'.join(parts[:i+1]))
    return ancestors

def extend_with_ancestors(labels):
    extended_set = set()
    for label in labels:
        extended_set.update(find_ancestors(label))
    return extended_set

def hierarchical_precision_recall_fmeasure(true_labels_series, predicted_labels_series, beta=1.0):
    # Convert Series to lists of single-element sets
    true_labels = [set([label]) for label in true_labels_series]
    predicted_labels = [set([label]) for label in predicted_labels_series]

    # Initialize accumulators for summing hP and hR across all instances
    sum_hP, sum_hR = 0, 0

    # Iterate over each pair of true and predicted label sets
    for true_set, pred_set in zip(true_labels, predicted_labels):
        extended_true = extend_with_ancestors(true_set)
        extended_pred = extend_with_ancestors(pred_set)

        # Calculate the correct predictions
        correct_predictions = extended_true.intersection(extended_pred)

        # Compute hierarchical precision and recall for the current instance
        hP = len(correct_predictions) / len(extended_pred) if extended_pred else 0
        hR = len(correct_predictions) / len(extended_true) if extended_true else 0

        sum_hP += hP
        sum_hR += hR

    # Calculate micro-averaged hP and hR
    micro_hP = sum_hP / len(true_labels) if true_labels else 0
    micro_hR = sum_hR / len(predicted_labels) if predicted_labels else 0

    # Calculate the F-measure
    hF = ((beta**2 + 1) * micro_hP * micro_hR / (beta**2 * micro_hP + micro_hR)) if (micro_hP + micro_hR) > 0 else 0

    return micro_hP, micro_hR, hF

def concatenate_csv_files(directory_path):
    # Initialize an empty list to store the dataframes
    dfs = []

    # Iterate over all the files in the directory
    for filename in os.listdir(directory_path):
        if filename.endswith('.csv'):
            # Construct the full file path
            file_path = os.path.join(directory_path, filename)
            # Read the CSV file into a dataframe
            df = pd.read_csv(file_path, index_col=0)
            # Rename the column to the filename without the extension
            df.columns = [os.path.splitext(filename)[0]]
            # Append the dataframe to the list
            dfs.append(df)

    # Concatenate all dataframes in the list horizontally
    concatenated_df = pd.concat(dfs, axis=1)

    return concatenated_df

def evaluate_models(predictions_df, y_val):
    # Initialize a dictionary to hold the results
    results = []

    # Iterate over each column (model) in the DataFrame
    for model in predictions_df.columns:
        # Extract the predictions for the current model
        y_pred = predictions_df[model].values

        # Calculate the metrics
        accuracy = accuracy_score(y_val, y_pred)
        macro_f1 = f1_score(y_val, y_pred, average='macro')
        mcc = matthews_corrcoef(y_val, y_pred)
        hP, hR, hF = hierarchical_precision_recall_fmeasure(y_val, y_pred)

        # Append the results
        results.append({
            'model': model,
            'accuracy': accuracy,
            'macro_f1': macro_f1,
            'mcc': mcc,
            'hP': hP,
            'hR': hR,
            'hF': hF
        })

    # Convert the results to a DataFrame
    results_df = pd.DataFrame(results)

    return results_df

def add_conflicts(df):
    for i, row in df.iterrows():
        # Find the index of the first None value
        none_indexes = row.isnull().argmax()
        first_none_index = none_indexes if none_indexes > 0 else len(row)

        # Define the range of columns to check for duplicates
        cols_to_check = df.columns[:first_none_index]

        # Extract the values to compare from the current row
        values_to_compare = row[cols_to_check]

        # Create a mask to identify rows with identical values up to the first None
        mask = (df[cols_to_check] == values_to_compare).all(axis=1) & (df.index != i)

        # Determine if there is any conflict
        df.loc[i, 'Conflict'] = mask.any()
        return df

def resolve_conflicts(dataframe):
    for i, row in dataframe.iterrows():
        # Check if the row is marked as a conflict
        if row['Conflict']:
            # Find the first None value and replace it with "End"
            for col in dataframe.columns:
                if pd.isnull(row[col]):
                    dataframe.at[i, col] = "."
                    break
    return dataframe

# Function to adjust values based on the depth (distance to the first None)
def adjust_values(row):
    # Find the index of the first None value in the row (ignoring the last three columns which are 'Values', 'Color', 'Conflict')
    first_none_index = next((i for i, x in enumerate(row[:-10]) if x is None), len(row))
    
    # Adjust the 'Values' based on the depth
    adjusted_value = row['Values'] / (first_none_index)
    
    return adjusted_value

def encode_fullstops(strings):
    # Initialize an empty list to hold the 1-hot encoded vector
    encoded_vector = []
    
    # Iterate over each string in the input list
    for item in strings:
        # Append 1 to the vector if the item is a full stop, otherwise append 0
        encoded_vector.append(0 if item == '.' else 1)
    
    return encoded_vector

def filter_elements(original_list, encoded_vector):
    # Ensure the two lists are of the same length
    if len(original_list) != len(encoded_vector):
        raise ValueError("Both lists must be of the same length.")
    
    # Use list comprehension to filter the original list based on the encoded vector
    filtered_list = [element for element, flag in zip(original_list, encoded_vector) if flag == 1]
    
    return filtered_list

def plot_single_line(results_df):
    df_melted = results_df.melt(id_vars=["model"], var_name="Metric", value_name="Score")

    # Plotting the data
    plt.figure(figsize=(12, 8))
    for metric in df_melted['Metric'].unique():
        subset = df_melted[df_melted['Metric'] == metric]
        plt.plot(subset['model'], subset['Score'], marker='o', label=metric)

    plt.xlabel('Model')
    plt.ylabel('Score')
    plt.title('Performance Metrics for Different Models')
    plt.xticks(rotation=90, ha='right')
    plt.legend(title='Metric')
    plt.tight_layout()

    plt.show()
    
def plot_multi_line(results_df):
    # Adjusting the figure for the correct number of metrics
    fig, axs = plt.subplots(n_metrics, 1, figsize=(12, 20), sharex=True)

    metrics = results_df.columns[1:]  # All metrics excluding the model names

    for i, metric in enumerate(metrics):
        axs[i].plot(results_df['model'], results_df[metric], marker='o', label=metric)
        axs[i].set_ylabel(metric)
        axs[i].legend(loc='upper left')
        axs[i].grid(True)

    # Set common labels and adjust layout
    plt.xlabel('Model')
    plt.xticks(rotation=45, ha='right')
    plt.suptitle('Performance Metrics for Different Models')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

def plot_accuracy_vs_frequency(y_preds_df, y_val):
    class_counts = y_val.value_counts().sort_values(ascending=False)

    model_accuracies = pd.DataFrame(index=y_val.unique())

    # Iterate over each model column (excluding 'Original_Targets' and 'Coverage')
    for model in y_preds_df.columns:
        accuracies = []
        for target in model_accuracies.index:
            # Select only the rows corresponding to the current class
            subset = y_preds_df[y_val == target]
            # Calculate the accuracy as the mean of correct predictions
            accuracy = (subset[model] == target).mean()
            accuracies.append(accuracy)

        # Assign the computed accuracies to the respective model column
        model_accuracies[model] = accuracies

    # Ensure the order of rows in model_accuracies matches the sorted class_counts
    # Assuming class_counts is defined and sorted as shown previously
    model_accuracies = model_accuracies.reindex(class_counts.index)

    # Ensure 'class_counts', 'model_accuracies', and 'model_colors' are defined
    fontsize_titles = 20
    fontsize_ticks = 16
    fontsize_legend = 16

    # Create a figure and a set of subplots
    fig, ax1 = plt.subplots(figsize=(15, 5))

    # Bar plot for class frequencies on the first y-axis
    bars = ax1.bar(class_counts.index, class_counts, color='#ededed', label='Class Frequency', width=0.8, align='edge')
    ax1.set_xlabel('Lineage', fontsize=fontsize_titles)
    ax1.set_ylabel('Frequency', fontsize=fontsize_titles)
    ax1.tick_params(axis='y', labelsize=fontsize_ticks)
    ax1.set_xticks([])  # Hide the class tick labels
    ax1.set_yscale('log')  # Set the y-axis to a logarithmic scale

    # Create the second y-axis to plot the model accuracies
    ax2 = ax1.twinx()
    ax2.set_ylabel('Accuracy', fontsize=fontsize_titles)
    ax2.tick_params(axis='y', labelsize=fontsize_ticks)

    # Plot each model's accuracy as separate lines and markers to control their alpha values individually
    for column in model_accuracies.columns:
        # Plot the line with one alpha value
        ax2.plot(model_accuracies.index, model_accuracies[column], linestyle='--', alpha=0.7, linewidth=2)
        # Overlay the markers with a different alpha value
        ax2.plot(model_accuracies.index, model_accuracies[column], label=column, marker='o', linestyle='', alpha=1, markersize=4)

    # Adjust layout and add legend outside the plot
    fig.tight_layout()
    legend = ax2.legend(title='Features', loc='lower left', bbox_to_anchor=(0, 1), borderaxespad=0., fontsize=16)
    plt.setp(legend.get_title(), fontsize=20)
    plt.savefig('Accuracy vs Frequency.png')
    plt.show()
```

```python
class_counts = y_val.value_counts().sort_values(ascending=False)

model_accuracies = pd.DataFrame(index=y_val.unique())

# Iterate over each model column (excluding 'Original_Targets' and 'Coverage')
for model in y_preds_df.columns:
    accuracies = []
    for target in model_accuracies.index:
        # Select only the rows corresponding to the current class
        subset = y_preds_df[y_val == target]
        # Calculate the accuracy as the mean of correct predictions
        accuracy = (subset[model] == target).mean()
        accuracies.append(accuracy)

    # Assign the computed accuracies to the respective model column
    model_accuracies[model] = accuracies

# Ensure the order of rows in model_accuracies matches the sorted class_counts
# Assuming class_counts is defined and sorted as shown previously
model_accuracies = model_accuracies.reindex(class_counts.index)

simple_lineages = pd.concat([model_accuracies, y_val.value_counts()], axis=1)
paths = [item for item in simple_lineages.index if item.count('.') <= 13]
split_strings = [s.split('.') + [None] * (13 - s.count('.')) for s in paths]

# Creating a DataFrame from the split strings
df = pd.DataFrame(split_strings, columns=['Level 1', 'Level 2', 'Level 3', 'Level 4', 'Level 5', 'Level 6', 'Level 7', 'Level 8', 'Level 9', 'Level 10', 'Level 11', 'Level 12', 'Level 13', 'Level 14'])
df['Values'] = np.log(simple_lineages["count"].tolist())
model_names = model_accuracies.columns

for name in model_names:
    df[name] = simple_lineages[name].tolist()

df = add_conflicts(df.copy())
df = resolve_conflicts(df.copy())
df['Values'] = df.apply(adjust_values, axis=1)

features = model_names
num_features = len(features)
```

```python
# Initialize a subplot figure with vertical stacking
fig = make_subplots(rows=num_features, cols=1, specs=[[{"type": "domain"}]] * num_features, vertical_spacing=0)

for i, feature in enumerate(features, start=1):
    # Update your DataFrame's color based on the feature
    df["Color"] = df[feature]

    # Create a Plotly Express sunburst figure for each feature
    px_fig = px.sunburst(df, path=['Level 1', 'Level 2', 'Level 3', 'Level 4', 'Level 5', 'Level 6', 'Level 7', 'Level 8', 'Level 9', 'Level 10', 'Level 11', 'Level 12', 'Level 13', 'Level 14'], values='Values', color="Color")
    
    px_fig.show()
    # Extract data for custom control
    labels = px_fig['data'][0]['labels'].tolist()
    parents = px_fig['data'][0]['parents'].tolist()
    values = px_fig['data'][0]['values'].tolist()
    ids = px_fig['data'][0]['ids'].tolist()
    colors = px_fig.data[0].marker.colors

    removes = encode_fullstops(labels)
    new_labels = filter_elements(labels, removes)
    new_parents = filter_elements(parents, removes)
    new_values = filter_elements(values, removes)
    new_ids = filter_elements(ids, removes)
    new_colors = filter_elements(colors, removes)

    # Add the customized sunburst trace to the subplot
    fig.add_trace(go.Sunburst(
        labels=new_labels,
        parents=new_parents,
        values=new_values,
        ids=new_ids,
        branchvalues='total',
        insidetextorientation='radial',
        marker=dict(
            colors=new_colors,  # your color values
            colorscale='Blues',
            cmin=0,  # setting minimum of color range
            cmax=1,  # setting maximum of color range
            showscale=(i == num_features)  # Show color scale only on the last plot
        )
    ), row=i, col=1)

# Update layout to fit the stacked sunburst plots without gaps
fig.update_layout(
    height=1000 * num_features,
    width=1000,
    margin=dict(t=0, l=0, b=0, r=0)
)

for i, feature in enumerate(features, start=1):
    fig.add_annotation(
        text=feature,  # Subplot title text
        xref="paper", yref="paper",
        x=0.5, y=1 - (i - 0.5) / num_features,
        xanchor="center", yanchor="bottom",
        showarrow=False,
        font=dict(size=14, color="white")  # Set text color to white here
    )

fig.write_image("yourfile.png") 
```
