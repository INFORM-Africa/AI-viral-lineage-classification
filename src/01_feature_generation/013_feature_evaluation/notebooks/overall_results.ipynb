```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, matthews_corrcoef
from sklearn.preprocessing import LabelEncoder
import itertools
import matplotlib.pyplot as plt
```

```python
# Function to calculate metrics
def calculate_metrics(y_true, y_pred):
    return {
        'accuracy': accuracy_score(y_true, y_pred),
        'micro_f1': f1_score(y_true, y_pred, average='micro'),
        'macro_f1': f1_score(y_true, y_pred, average='macro'),
        'matthews_corrcoef': matthews_corrcoef(y_true, y_pred)
    }

# Function to handle the expansion for a single cell in the DataFrame
def expand_cell(lineage):
    # Here we assume that expand_aliases returns a list and we just want the first item
    # Adjust this as needed based on how you want to handle multiple possible expansions
    expanded = expand_aliases(lineage, mappings)
    return expanded[0] if expanded else lineage

# Function to calculate the Longest Common Ancestor metric
def calculate_lca(target, prediction):
    target_levels = target.split('.')
    prediction_levels = prediction.split('.')
    min_length = min(len(target_levels), len(prediction_levels))
    
    # Count the number of matching levels from the start
    matching_levels = 0
    for i in range(min_length):
        if target_levels[i] == prediction_levels[i]:
            matching_levels += 1
        else:
            break  # Stop at the first non-matching level
    
    # LCA metric can be the number of matching levels divided by the total number of levels in the target
    lca_metric = matching_levels / len(target_levels)
    return lca_metric

def expand_aliases(lineage, mappings, depth=0, max_depth=10):
    # Preventing infinite recursion
    if depth > max_depth:
        print(f"Reached max depth with lineage: {lineage}")
        return [lineage]
    
    # Split the lineage into parts based on the dot separator
    parts = lineage.split('.')
    
    # Initialize a list to hold all possible expanded lineage combinations
    expanded_combinations = ['']
    
    for part in parts:
        new_combinations = []
        if part in mappings:
            # If the mapping is a list, iterate through all options
            if isinstance(mappings[part], list):
                for option in mappings[part]:
                    # Append each option to each existing combination
                    for combo in expanded_combinations:
                        new_combinations.append(combo + (combo and '.') + option)
            else:
                # Single mapping, append to each existing combination
                for combo in expanded_combinations:
                    new_combinations.append(combo + (combo and '.') + mappings[part])
        else:
            # No mapping, append the part to each existing combination
            for combo in expanded_combinations:
                new_combinations.append(combo + (combo and '.') + part)
        
        # Update the expanded combinations with the new ones
        expanded_combinations = new_combinations
    
    return expanded_combinations

def get_results(mappings, le, params):
    predictions = pd.read_csv(f"../predictions/SARS/{params.get('model')}_{params.get('p_set')}.csv")
    predictions = predictions.drop(columns=["FCGR_remove_128"])
    column_mapping = {
        '7-mer_FFP_remove': 'FFP',
        '7-mer_remove': 'K-mer',
        '7-spaced_remove': 'Spaced',
        'ACS_remove': 'ACS',
        'DSP_replace_real': 'GSP',
        'FCGR_remove_256': 'FCGR',
        'Mash_distance_remove_21': 'Mash',
        'RTD_7-mer_remove': 'RTD'
    }

    predictions = predictions.rename(columns=column_mapping)
    predictions['Original_Targets_Encoded'] = le.transform(predictions['Original_Targets'])
    
    # Initialize an empty list to store result rows
    result_rows = []

    # Iterate through each prediction column to calculate metrics
    for column in predictions.columns[1:-1]:  # Skipping the 'Original_Targets' and 'Original_Targets_Encoded' columns
        # Transform prediction column to numerical values
        y_pred = le.transform(predictions[column])
        # Calculate metrics
        metrics = calculate_metrics(predictions['Original_Targets_Encoded'], y_pred)
        # Add the results to the list
        result_rows.append({
            'Model': params.get('model'),
            'Set': params.get('p_set'),
            'Feature': column,
            'Accuracy': metrics['accuracy'],
            'Micro F1': metrics['micro_f1'],
            'Macro F1': metrics['macro_f1'],
            'Matthews CorrCoef': metrics['matthews_corrcoef']
        })

    # Convert the list of result rows into a DataFrame
    results_df = pd.DataFrame(result_rows)
    
    expanded_df = predictions.drop(columns="Original_Targets_Encoded").applymap(expand_cell)
    
    lca_results = {}
    for column in expanded_df.columns[1:]:  # Skip 'Original_Targets' column
        lca_results[column] = expanded_df.apply(lambda row: calculate_lca(row['Original_Targets'], row[column]), axis=1)

    # Convert the LCA results to a DataFrame for easier viewing/manipulation
    lca_df = pd.DataFrame(lca_results)

        # Calculate the average LCA metric for each column
    average_lca = lca_df.mean()
    results_df["LCA"] = average_lca.values
    return results_df
```

```python
import json

# Assuming the JSON file is named 'mappings.json'
with open('alias_key.json', 'r') as f:
    mappings = json.load(f)

data = pd.read_parquet('../../../data/processed/cov-19.parquet', engine='pyarrow')
data = data.drop(columns = "Sequence")

le = LabelEncoder()
le.fit(data["Lineage"])
```

```python
# Define the options for each parameter
options = {
    "model": ["random_forest", "xgboost", "knn"],
    "p_set": ["Training", "Validation", "Testing"]
}

# Use itertools.product to generate all possible combinations
parameter_combinations = list(itertools.product(*options.values()))

# Convert each combination to a dictionary and create a list of dictionaries
parameter_grid = [dict(zip(options.keys(), combination)) for combination in parameter_combinations]

frames = []
# Display the parameter grid
for params in parameter_grid:
    frames.append(get_results(mappings, le, params))

results_table = pd.concat(frames)
```

```python
results_table = results_table[results_table["Feature"] != "FCGR_remove_128"]
```

```python
results_table[results_table["Set"] == "Testing"].drop(columns="Set").round(4)
```

```python
# Pivot the dataframe to format suitable for plotting
pivot_df = results_table[results_table["Set"]=="Testing"].pivot(index='Feature', columns='Model', values='Accuracy')

# Plotting
# Plotting as a horizontal bar plot
pivot_df.plot(kind='barh', figsize=(6, 6))
plt.title('Model Accuracy by Feature (Testing Set)')
plt.xlabel('Accuracy')
plt.ylabel('Feature')
plt.legend(title='Model')
plt.tight_layout()

plt.show()
```

```python
# Reordering the 'Set' column to ensure the correct order in the plots
set_order = ['Testing', 'Validation', 'Training']
results_table['Set'] = pd.Categorical(results_table['Set'], categories=set_order, ordered=True)

blue_shades = ['#8cc5e3', '#2066a8', '#3594cc']
three_colour = ['#ea801c', '#b8b8b8', '#1a80bb']

# Setting up the plot with 3 subplots, one for each model
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

models = results_table['Model'].unique()
for i, model in enumerate(models):
    # Filtering and pivoting data for each model
    model_data = results_table[results_table['Model'] == model].sort_values('Set')
    pivot_df = model_data.pivot(index='Feature', columns='Set', values='Accuracy')

    # Plotting with the specified bar order and color scheme
    pivot_df.plot(kind='barh', ax=axes[i], width=0.8, color=three_colour)
    axes[i].set_title(f'{model.capitalize()} Model Accuracy by Set')
    axes[i].set_xlabel('Accuracy')

    # Setting y-axis label and tick labels only for the first subplot
    if i == 0:
        axes[i].set_ylabel('Feature')
    else:
        axes[i].set_ylabel('')
        axes[i].set_yticklabels([])  # Hide y-axis tick labels for the second and third subplots

    axes[i].legend(title='Set')

plt.tight_layout()
plt.show()
```
