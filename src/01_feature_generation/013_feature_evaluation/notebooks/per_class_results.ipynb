```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, matthews_corrcoef
from sklearn.preprocessing import LabelEncoder
import itertools
import matplotlib.pyplot as plt
```

```python
data = pd.read_parquet('../../../data/processed/cov-19.parquet', engine='pyarrow')
data = data.drop(columns = "Sequence")
data = data[data["Train"] == 2]
```

```python
predictions = pd.read_csv('../predictions/SARS/random_forest_Testing.csv')
predictions = predictions.drop(columns=["FCGR_remove_128"])
```

```python
predictions['Coverage'] = data['Coverage'].tolist()
```

```python
df = predictions
```

```python
column_mapping = {
    '7-mer_FFP_remove': 'FFP',
    '7-mer_remove': 'K-mer',
    '7-spaced_remove': 'Spaced',
    'ACS_remove': 'ACS',
    'DSP_replace_real': 'GSP',
    'FCGR_remove_256': 'FCGR',
    'Mash_distance_remove_21': 'Mash',
    'RTD_7-mer_remove': 'RTD'
}

df = df.rename(columns=column_mapping)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

class_counts = df['Original_Targets'].value_counts().sort_values(ascending=False)

model_accuracies = pd.DataFrame(index=df['Original_Targets'].unique())

# Iterate over each model column (excluding 'Original_Targets' and 'Coverage')
for model in df.columns.drop(['Original_Targets', 'Coverage']):
    accuracies = []
    for target in model_accuracies.index:
        # Select only the rows corresponding to the current class
        subset = df[df['Original_Targets'] == target]
        # Calculate the accuracy as the mean of correct predictions
        accuracy = (subset[model] == target).mean()
        accuracies.append(accuracy)
    
    # Assign the computed accuracies to the respective model column
    model_accuracies[model] = accuracies

# Ensure the order of rows in model_accuracies matches the sorted class_counts
# Assuming class_counts is defined and sorted as shown previously
model_accuracies = model_accuracies.reindex(class_counts.index)
```

```python
# model_colors = ['#dd571c', '#ff7900', '#ffa500', '#fbceb1', '#082a54', '#007191', '#008dff', '#8fd7d7']  # Define your color list here
model_colors = ['#bdd373', '#98c127', '#ff8ca1', '#f45f74', '#00b0be', '#8fd7d7', '#f0c571', '#ff9d3a']  # Define your color list here
```

```python
import matplotlib.pyplot as plt

# Ensure 'class_counts', 'model_accuracies', and 'model_colors' are defined
fontsize_titles = 20
fontsize_ticks = 16
fontsize_legend = 16

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(15, 5))

# Bar plot for class frequencies on the first y-axis
bars = ax1.bar(class_counts.index, class_counts, color='#ededed', label='Class Frequency', width=0.8, align='edge')
ax1.set_xlabel('Lineage', fontsize=fontsize_titles)
ax1.set_ylabel('Frequency', fontsize=fontsize_titles)
ax1.tick_params(axis='y', labelsize=fontsize_ticks)
ax1.set_xticks([])  # Hide the class tick labels
ax1.set_yscale('log')  # Set the y-axis to a logarithmic scale

# Create the second y-axis to plot the model accuracies
ax2 = ax1.twinx()
ax2.set_ylabel('Accuracy', fontsize=fontsize_titles)
ax2.tick_params(axis='y', labelsize=fontsize_ticks)

# Plot each model's accuracy as separate lines and markers to control their alpha values individually
for column, color in zip(model_accuracies.columns, model_colors):
    # Plot the line with one alpha value
    ax2.plot(model_accuracies.index, model_accuracies[column], color=color, linestyle='--', alpha=0.7, linewidth=2)
    # Overlay the markers with a different alpha value
    ax2.plot(model_accuracies.index, model_accuracies[column], label=column, marker='o', linestyle='', color=color, alpha=1, markersize=4)

# Adjust layout and add legend outside the plot
fig.tight_layout()
legend = ax2.legend(title='Features', loc='upper left', bbox_to_anchor=(1.10, 1), borderaxespad=0., fontsize=16)
plt.setp(legend.get_title(), fontsize=20)
plt.savefig('Accuracy vs Frequency.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming 'df' is your DataFrame
sorted_df = df.sort_values(by='Coverage', ascending=False)
coverage_values = sorted_df['Coverage']

# Define the bin edges for coverage values
bin_edges = np.arange(95, 100.1, 0.1)  # Bins adjusted as needed

# Use pd.cut to bin the coverage data
sorted_df['CoverageBin'] = pd.cut(sorted_df['Coverage'], bins=bin_edges, include_lowest=True, right=False)

# Calculate the frequency of lineages per bin
coverage_freq = sorted_df['CoverageBin'].value_counts().sort_index()

# Convert index to the midpoints of each bin
bin_mids = [interval.mid for interval in coverage_freq.index.categories]

# Initialize an empty DataFrame for average accuracies for each model per bin
model_accuracies = pd.DataFrame(index=coverage_freq.index.categories, columns=sorted_df.columns.drop(['Original_Targets', 'Coverage', 'CoverageBin']))

# Calculate the average accuracy for each model in each bin
for bin_interval in coverage_freq.index.categories:
    bin_subset = sorted_df[sorted_df['CoverageBin'] == bin_interval]
    for model in model_accuracies.columns:
        correct_predictions = bin_subset.apply(lambda x: x[model] == x['Original_Targets'], axis=1)
        model_accuracies.at[bin_interval, model] = correct_predictions.mean()

# Create the plot
fig, ax1 = plt.subplots(figsize=(15, 5))

# Bar plot for lineage frequency based on binned coverage
ax1.bar(bin_mids, coverage_freq, width=np.diff(bin_edges)[0] * 0.8, color='#ededed', align='center')
ax1.set_xlabel('Coverage', fontsize=fontsize_titles)
ax1.set_ylabel('Frequency of Lineages', fontsize=fontsize_titles)
ax1.tick_params(axis='y', labelsize=fontsize_ticks)
ax1.tick_params(axis='x', labelsize=fontsize_ticks)
ax1.invert_xaxis()  # Flip the x-axis to show higher coverage values on the left

# Assuming 'model_colors' list is defined with enough colors for each model

# Create the second y-axis to plot the model accuracies
ax2 = ax1.twinx()
ax2.set_ylabel('Accuracy', fontsize=fontsize_titles)
ax2.tick_params(axis='y', labelsize=fontsize_ticks)

# Plot each model's averaged accuracy as separate lines
for column, color in zip(model_accuracies.columns, model_colors):
    ax2.plot(bin_mids, model_accuracies[column], label=column, marker='o', linestyle='-', color=color, alpha=0.7, markersize=8)

# Adjust layout and add legend outside the plot
fig.tight_layout()
legend = ax2.legend(title='Features', loc='upper left', bbox_to_anchor=(1.10, 1), borderaxespad=0., fontsize=fontsize_legend)
plt.setp(legend.get_title(), fontsize=fontsize_titles)
plt.savefig('Accuracy vs Coverage.png')
plt.show()
```
